## The set() Constructor

```python
myset = set(('katrushka','katy','adelita'))
print(myset)
```

## Change Items


Once a set is created, you cannot change its items, but you can add new items.


```python
l = list(myset)
print(l) 
l[1] = 'gaty'
myset = set(l) 
print(myset)
```

## Add set items

```python
print(myset)
```

```python
myset.add('katy')
print(myset)
```

## Update sets
To add items from another set into the current set, use the **update()** method.

```python
myset = {'red','green','blue'}
sColors = {'cyan','yellow'}
myset.update(sColors)
print(myset)
```

## Add Any Iterable
The object in the **update()** method **does not have to be a set, it can be any iterable object** (tuples, lists, dictionaries etc.).


```python
print(myset)
```

```python
setFig = ['square','circle','triangle']
```

```python
myset.update(setFig)
print(myset)
```

# Remove set items
## Remove Item
To remove an item in a set, use the **remove()**, or the **discard()** method.

```python
myset = {'red', 'square', 'blue', 'triangle', 'yellow', 'green','cyan'}
```



**Note: If the item to remove does not exist, discard() will NOT raise an error.**


```python
myset.discard('circle')
print(myset)
```

```python
myset.discard('circle')
print(myset)
```

```python
myset.remove('cyan')
print(myset)
```

```python
myset.remove('rosa')
print(myset)
```

You can also use the **pop()** method to remove an item, but this method will remove the last item. Remember that sets are unordered, so you will not know what item that gets removed.

```python
print(myset)
```

The **clear()** method empties the set:

```python
myset.clear()
print(myset)
```

# Join sets
There are several ways to join 2 or more sets in Python.

The **union()** method returns a new set with all items from both sets:

```python
set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}
set3 = set1.union(set2)
print(set3) 
```

The **update()** method inserts the items in set2 into set1:


```python
set2.update(set1)
print(set2)
```



**Note: Both union() and update() will exclude any duplicate items.**


```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}  
x.update(y) 
print(x)
```

## Keep ONLY the Duplicates
The **intersection_update()** method will keep only the items that are present in both sets.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.intersection_update(y)
```

```python
print(x)
```

The **intersection()** method will return a new set, that only contains the items that are present in both sets.

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.intersection(y)
print(x.intersection(y))
print(x)
```

```python
z = x.intersection(y)
print(z)
```

## Keep All, But NOT the Duplicates
The **symmetric_difference_update()** method will keep only the elements that are NOT present in both sets.

```python
x.symmetric_difference_update(y)
print(x)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
z = x.symmetric_difference(y)
print(z)
```
