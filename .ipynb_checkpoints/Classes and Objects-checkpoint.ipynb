A **Class** is like an object constructor, or a "blueprint" for creating objects.

## Create a Class

```python
class MyClass:
    value = 'Annie'
```

## Create an object

```python
obj1 = MyClass()
print(obj1.value)
```

## The __ init __() Function
All classes have a function called **__ init __()**, which is always executed when the class is being initiated.

Use the __ init __() function to assign values to object properties, or other operations that are necessary to do when the object is being created.

The **self** parameter is a reference to the current instance of the class, and is used to access variables that belong to the class.

**Example**: Create a class named Person, use the __ init __() function to assign values for fName,lName and age

```python
class Person:
    def __init__(self,fName,lName,age):
        self.fName = fName
        self.lName = lName
        self.age = age
        
person = Person('George','Orwell',34)
print(person.age)
```

## Object methods
Methods in objects are functions that belong to the object.

```python
class Person:
    def __init__(self,fName,lName,age):
        self.fName = fName
        self.lName = lName
        self.age = age
    def funcPerson(self):
        print('Hi, I\'m ' + self.fName)
```

```python
person2 = Person('Winston','Smith',42)
person2.funcPerson()
```

## Modify object properties

```python
person2.age = 4
print(person2.age)
```

## Delete Object Properties

```python
del person2.age
```

## Delete Objects

```python
del person2
```
