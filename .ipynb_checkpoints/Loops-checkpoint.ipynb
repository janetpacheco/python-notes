# While loop
Python has two primitive loop commands:
* while loops
* for loops


**Example** 

Print i as long as i is less than 6:

```python
i = 0
while i < 6:
    print(i)
    i+=1
```

##  The break statement

Exit the loop when is 3

```python
i = 1
while i < 6:
    print(i)    
    if i == 3:
        break
    i += 1
    
    
```

## The continue statement
With the **continue** statement we can stop the current iteration, and continue with the next:

**Example:**

Continue to the next iteration if i is 3:

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

## The else statement
With the **else** statement we can run a block of code once when the condition no longer is true

```python
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print('i is no longer less than 6')
```

# For loops

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    print(x)
```

## Looping Through a String

```python
for x in "banana":
    print(x)
```

## The break statement

```python
fruits = ['apple','pi単a','cherry']
for f in fruits:
    print(f)
    if f == 'pi単a':
        break
```

```python
for f in fruits:
    if f =='pi単a':
        break
    print(f)
```

## The continue statement

```python
countries = ['Germany','Turkey','Italy','Rusia']
for c in countries:
    if c == 'Turkey':
        continue
    print(c)    
    
```

## The range() function

```python
for i in range(5):
    print(i)
```

```python
for i in range(3,7):
    print(i) 
```

```python
for x in range(5,50,5):
    print(x)
```

## Else in For Loop

```python
for m in range(10):
    print(m)
else:
    print('Finally finished')
```

Break the loop when **x** is 3, and see what happens with the else block:

```python
for x in range(5):
    if x == 3:
        break
    print(x)
else:
    print('Finally finished')
```

## Nested loops

```python
adj = ['red','big','tasty']
fruits = ['apple','watermelon','guayaba']

for f in fruits:
    for a in adj:
        print(f,a)
       
```

## Pass statement

```python
for num in [1,2,3]:
    pass
```
