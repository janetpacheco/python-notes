## Check if key exists
To determine if a specified key is present **in** a dictionary use the in keyword:

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
if 'model' in thisdict:
    print('Yes,"model" is one of the keys in thisdict dictionary')
```

## Remove dictionary items
There are several methods to remove items from a dictionary:

```python
print(thisdict)
```

```python
thisdict.pop('model')
print(thisdict)
```

The **popitem()** method removes the last inserted item (in versions before 3.7, a random item is removed instead):

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
```

The **clear()** method empties the dictionary:

```python
thisdict.clear()
print(thisdict)
```

## Loop dictionaries

```python
print(myDict)
```

**Print all key** names in the dictionary, one by one:

```python
for element in myDict:
    print(element)
```

You can use the **keys()** method to return the keys of a dictionary:

```python
for element in myDict.keys():
    print(element) 
```

**Print all values** in the dictionary, one by one:

```python
for x in myDict:
    print(myDict[x])
```

You can also use the **values()** method to return values of a dictionary:

```python
for element in myDict.values():
    print(element)
```

Loop through both keys and values, by using the **items()** method:

```python
for x, y in myDict.items():
    print(x, y)
```

## Copy dictionaries
Make a copy of a dictionary with the **copy()** method:

```python
print(myDict)
```

```python
myCopy = myDict.copy()
print(myCopy)
```

```python
myCopy['model']= 'Nissan'
print(myCopy)
print(myDict)
```

Make a copy of a dictionary with the **dict()** function:

```python
myCopy2 = dict(myDict)
print(myCopy2)
```

## Nested dictionaries
Create a dictionary that contain three dictionaries:

```python
myPets ={
    'perrito1':{
        'name':'Adelita',
        'year': 2019        
    },
    
    'perrito2':{
        'name':'Masha',
        'year': 2018        
    },
    
    'perrito3':{
        'name':'Costelo',
        'year': 2017        
    }
}
```

```python
print(myPets)
```

```python
child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
} 
```

```python
print(myfamily)
```
