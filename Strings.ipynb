# Strings

## Multiline strings
You can assign a multiline string to a variable by using three quotes:

```python
a = """Years of separation cannot withhold us .. 
Hearts dispute distance can connect .. 
A heart cannot forget a lover who was once adored .. 
And the stars can not separate from space ."""
print(a) 
```

## Strings are arrays

**Slicing notation**

```python
s = 'conejo'
print(s[0]) 
```

## Looping through a string

```python
for c in 'conejita':
    print(c) 
```

## String length

```python
c = 'conejita'
print(len(c))
```

## Check String

```python
txt = 'The best things in life are free!'
print("free" in txt)
```

Use it in a if statement

```python
txt = "The best things in life are free!"
if "free" in txt:
    print("Yes, 'free' is present.")
```

## Check if NOT

```python
txt = "The best things in life are free!"
print('expensive'not in txt)
```

```python
txt = "The best things in life are free!"
if "amiga" not in txt:
    print("No, 'amiga' is not present.")
```

## Slicing

```python
phrase = "Adelita de mi corazon"
```

```python
print(phrase[0:4])
```

### From start

```python
print(phrase[:4])
```

### To the end

```python
print(phrase[1:])
```

### Negative indexing

```python
print(phrase[-2:])
```

From 'c' to, but not included: 'z'

```python
print(phrase[-7:-3])
```

## Modify strings
### Upper case

```python
p = '\'on tas?'
print(p.upper())
```

### Lower case

```python
p = 'FOOLS TALK, COWARDS ARE SILENT, WISE MEN LISTEN'
print(p.lower())
```

### Remove white space
Removes any whitespace from the beginning or the end.

```python
c ='     Memories    are worse than bullets.    '
print(c.strip())
```

### Replace string

```python
daniel = """Never trust anyone, Daniel, especially the people you admire. 
Those are the ones who will make you suffer the worst blows."""

print(daniel.replace('Daniel','Annie'))
```

### Split string

```python
a = 'Ich vermisse dich, aufrichtig dein Kaninchen.'
print(a.split(","))
```

### Conatenation

```python
x = 'Onta '
y = ' bebé'
z = '?'
full = x + y +z
print(full)
```

### Format
We cannot combine strings and numbers like this:

```python
age = 28
txt = 'I\'m Annie,I\'m' + age
print(txt)
```

```python
txt =  'I\'m Annie,I\'m {}' 
print(txt.format(age))
```

```python
quantity = 4
noItem = 45
price = 7
myOrder = 'I want {} pieces of item {} for {} pesos.'
print(myOrder.format(quantity,noItem,price))
```

```python
myOrder = 'I want {2} pieces of item {1} for {0} pesos.'

print(myOrder.format(quantity,noItem,price))
```

### Escape character

```python
txt = "We are the so-called \"Vikings\" from the north." 
print(txt)
```

```python
txt = "This will insert one \\ (backslash)."
print(txt) 
```

```python
txt = "\tWorld! Hello" 
print(txt) 
```
